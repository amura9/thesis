```python
import importlib
```

```python
def load_plugins(plugin_paths):
    plugins = []
    for path in plugin_paths:
        module_path, class_name = path.rsplit(".", 1)
        module = importlib.import_module(module_path)
        cls = getattr(module, class_name)
        plugins.append(cls())
    return plugins
```

#### In pratica prende ogni plugin indicato nel config, importa il file e crea un'istanza della classe!
