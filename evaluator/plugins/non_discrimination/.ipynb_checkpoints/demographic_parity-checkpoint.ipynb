```python
import numpy as np
```

```python
class DemographicParity:
    def __init__(self):
        self.name = "Demographic Parity"

    def evaluate(self, y_true, y_pred, sensitive_feature):
        groups = np.unique(sensitive_feature)
        rates = []
        
        for group in groups:
            idx = sensitive_feature == group
            if np.sum(idx) > 0:
                rate = np.mean(y_pred[idx])
                rates.append(rate)
        
        return 1 - np.abs(rates[0] - rates[1])  # più vicino a 1 è, meglio è
```
